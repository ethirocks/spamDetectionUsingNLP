```python
def remove_punc(text):
    text_nonpunct = "".join([char for char in text if char not in string.punctuation])
    return text_nonpunct

data['body_text_clean'] = data['body_text'].apply(lambda x: remove_punc(x))

data.head()
```

```python
def tokenize(text):
    tokens = re.split('\W+', text)
    return tokens

data['body_text_tokenized'] = data['body_text_clean'].apply(lambda x: tokenize(x.lower()))

data.head()
```

```python
stopword = nltk.corpus.stopwords.words('english')

def remove_stopwords(tokenized_list):
    text = [word for word in tokenized_list if word not in stopword]
    return text

data['body_text_nostop']= data['body_text_tokenized'].apply(lambda x: remove_stopwords(x))

data.head()
```

```python
ps = nltk.PorterStemmer()

def stemming(tokenized_text):
    text = [ps.stem(word) for word in tokenized_text]
    return text

data['body_text_stemmed'] = data['body_text_nostop'].apply(lambda x : stemming(x))

data.head()
```

```python
wn= nltk.WordNetLemmatizer()

def lemmatizing(tokenized_text):
    text = [wn.lemmatize(word) for word in tokenized_text]
    return text

data['body_text_lemmatize'] = data['body_text_nostop'].apply(lambda x : lemmatizing(x))

data.head()
```

```python
def clean_text(text):
    text = "".join([word.lower() for word in text if word not in string.punctuation])
    tokens = re.split('\W+', text)
    text = [ps.stem(word) for word in tokens if word not in stopword]
    return text
```

```python
count_vect = CountVectorizer(analyzer=clean_text)

X_counts =  count_vect.fit_transform(data['body_text'])

X_counts_df = pd.DataFrame(X_counts.toarray())

X_counts_df.columns = count_vect.get_feature_names()

X_counts_df
```

```python
def clean_text_two(text):
    text = "".join([word.lower() for word in text if word not in string.punctuation])
    tokens = re.split('\W+', text)
    text = " ".join([ps.stem(word) for word in tokens if word not in stopword])
    return text

data['cleaned_text'] = data['body_text'].apply(lambda x: clean_text_two(x))
```

```python
ngram_vect = CountVectorizer(ngram_range=[1,2])

X_counts_ngram = ngram_vect.fit_transform(data['cleaned_text'])

X_counts_ngram_df = pd.DataFrame(X_counts_ngram.toarray())

X_counts_ngram_df.columns = ngram_vect.get_feature_names()

X_counts_ngram_df
```

```python
tfidf_vect = TfidfVectorizer(analyzer = clean_text)
X_tfidf = tfidf_vect.fit_transform(data['body_text'])

X_counts_tfidf_df = pd.DataFrame(X_tfidf.toarray())

X_counts_tfidf_df.columns = tfidf_vect.get_feature_names()

X_counts_tfidf_df
```

```python
def count_punct(text):
    count = sum([1 for char in text if char in string.punctuation])
    return round(count/(len(text) - text.count(" ")),3)*100

data['body_len'] = data['body_text'].apply(lambda x: len(x) - x.count(" "))
data['punct%'] = data['body_text'].apply(lambda x: count_punct(x))

X_features = pd.concat([data['body_len'], data['punct%'], pd.DataFrame(X_tfidf.toarray())], axis=1)
X_features.head()
```

```python
def train_RF(n_est, depth):
    rf = RandomForestClassifier(n_estimators=n_est, max_depth=depth, n_jobs=-1)
    rf_model = rf.fit(X_train, y_train)
    y_pred = rf_model.predict(X_test)
    precision, recall, fscore, support = score(y_test, y_pred, pos_label='spam', average='binary')
    print('Est: {} / Depth: {} ---- Precision: {} / Recall: {} / Accuracy: {}'.format(
        n_est, depth, round(precision, 3), round(recall, 3),
        round((y_pred==y_test).sum() / len(y_pred), 3)))
```

```python
for n_est in [10, 50, 100]:
    for depth in [10, 20, 30, None]:
        train_RF(n_est, depth)
```
